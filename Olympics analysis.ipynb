```python
top_10_countries= athletes.Team.value_counts().sort_values(ascending= False).head(10)
top_10_countries
```

We can have a look on top 10 countries participated since the inception of Olympics.

```python
plt.figure(figsize= (15,6))
plt.title("Top ten countries")
sns.barplot(x= top_10_countries.index, y= top_10_countries, palette= 'inferno')
plt.xticks(rotation= 20)
plt.show()
```

```python
plt.figure(figsize= (15,6))
plt.title("Age distribution")
plt.xlabel('Age')
plt.ylabel('No.of participants')
plt.hist(athletes.Age, bins= np.arange(10,80,2), color= 'orange', edgecolor= 'white')
plt.show()
```

 Observations:

1. From the plot we can observe that the athletes with age group between 20-30 are highly participated in Olympics.
2. Interestingly we can also find some athletes who are beyond 40 years of age.

```python
gender= athletes.Sex.value_counts()
gender
```

```python
labels = athletes['Sex'].value_counts().index
size = athletes['Sex'].value_counts().values
plt.figure(figsize = (6,6))
plt.pie(size,colors = ['lightskyblue', 'gold'],explode = (0, 0.050), labels = labels,shadow = True,autopct = "%.2f%%")
plt.title('Male athletes v/s Female athletes')
plt.axis('off')
plt.legend()
plt.show()
```

```python
Medals= athletes.Medal.value_counts()
Medals
```

```python
women_olympics= athletes[(athletes.Sex == 'F') & (athletes.Season == 'Summer')]
women_olympics.head()
```

```python
plt.figure(figsize= (15,6))
plt.title("Female participants in Olympics")
sns.countplot(x= 'Year', data= women_olympics, palette= 'Spectral')
plt.xticks(rotation= 20)
plt.show()
```

Observations:

We can find the details and info of female athletes who are participated in Olympics. 

```python
part= women_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize= (15,6))
plt.title("Plot of participation by female in Olympics over time period.")
part.loc[:, 'F'].plot()
plt.show()
```

Observations:

We can observe the gradual increase in female participation. 

```python
gold_medals= athletes[athletes.Medal =='Gold']
gold_medals
```

```python
gold_medals= gold_medals[np.isfinite(gold_medals['Age'])]
gold_medals['ID'][gold_medals['Age'] > 60].count()
```

Observations:

Surprisingly there are some athletes whose age group is around 60.

```python
sporting_events= gold_medals['Sport'][gold_medals['Age'] > 60]
sporting_events
```

```python
plt.figure(figsize= (15,6))
plt.tight_layout()
plt.title("Gold medals for athletes over 60 years of age")
sns.countplot(sporting_events, palette='Set2')
plt.show()
```

```python
# Rio Olympics 

max_year= athletes.Year.max()
print(max_year)
```

```python
team_names= athletes[(athletes.Year == max_year) & (athletes.Medal == "Gold")].Team
team_names.value_counts().head()
```

```python
plt.figure(figsize= (15,6))
sns.barplot(x= team_names.value_counts().head(20), y= team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel("Country wise medals in Rio Olymics 2016")
plt.show()
```

```python
non_null_medals= athletes[(athletes['Height'].notnull()) & (athletes['Weight'].notnull())]
```
